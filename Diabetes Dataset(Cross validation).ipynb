# Data is imbalance 

```python
# Data balance by using RandomOver sampling technique
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
r=RandomOverSampler()
x_resample,y_resample=r.fit_resample(x,y)
print(Counter(y_resample))
```

```python
# Cross Validation technique
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=11,shuffle=True)
kf.get_n_splits(x)
print(kf)
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegression
l1=LogisticRegression()
l2=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=100)
```

```python
print(Counter(y_test))
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
l1.fit(x_train,y_train)
```

```python
l1.fit(x_train,y_train)
y_pred=l1.predict(x_test)
print(y_pred)
```

```python
scores=cross_val_score(l1,x_train,y_train,cv=kf)
print(scores)
print(np.mean(scores)*100)
```

```python
for train_data,test_data in kf.split(x):
    x_train,x_test=x_resample[train_data],x_resample[test_data]
    y_train,y_test=y_resample[train_data],y_resample[test_data]
    scores=cross_val_score(l1,x_train,y_train,cv=kf)
    print(scores)
    print("y data Counter: ",Counter(y_test))
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5,shuffle=True, random_state=21)
skf.get_n_splits(x_resample,y_resample)
```

```python
for train_data,test_data in skf.split(x_resample,y_resample):
    x_train,x_test=x_resample[train_data],x_resample[test_data]
    y_train,y_test=y_resample[train_data],y_resample[test_data]
    scores=cross_val_score(l1,x_train,y_train,cv=skf)
    y_pred=cross_val_predict(l1,x_test,y_test,cv=skf)
    print("Checking y_test Balancing:",Counter(y_test))
    print("Checking y_train Balancing:",Counter(y_train))
    print(scores)
    print(y_pred)
```
